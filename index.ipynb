```python
import serial
ad1=serial.Serial('com7',9600) #forearm
ad2=serial.Serial('com10',9600) #muscle
```

```python
ad1.flush()
ad2.flush()
```

```python
def getval1():
    while (ad1.inWaiting()==0):
        pass
    as1=str(ad1.readline())
    l=len(as1)
    as1=as1[2:l-5]
    #print(str(as1))
    return(str(as1))

def getval2():
    while (ad2.inWaiting()==0):
        pass
    as2=str(ad2.readline())
    l=len(as2)
    as2=as2[2:l-5]
    #print(str(as2))
    return(str(as2))
```

```python
ad1.flush()
ad2.flush()
```

```python
def func_incal1(tt):
    x=0
    y=0
    z=0
    m=0
    for i in range(0,tt):
        arr2=list(map(int,getval1().split(' ')))
        x+=arr2[0]
        y+=arr2[1]
        z+=arr2[2]
        m+=arr2[3]
        #print(m)
    #print("x : ",sum(x)/tt,"  y : ",sum(y)/tt,"  z : ",sum(z)/tt)
    return(x/tt,y/tt,z/tt,m/tt)



def func_incal2(tt):
    x=0
    y=0
    z=0
    m=0
    for i in range(0,tt):
        arr2=list(map(int,getval2().split(' ')))
        x+=arr2[0]
        y+=arr2[1]
        z+=arr2[2]
        m+=arr2[3]
        #print(m)
    #print("x : ",sum(x)/tt,"  y : ",sum(y)/tt,"  z : ",sum(z)/tt)
    return(x/tt,y/tt,z/tt,m/tt)

ad1.flush()
ad2.flush()

x1,y1,z1,m1=func_incal1(20)
x2,y2,z2,m2=func_incal2(20)

   
def func_incal12(tt):
    xx,yy,zz,mm=func_incal1(tt)
    return(xx-x1,yy-y1,zz-z1,mm)
def func_incal22(tt):
    xx,yy,zz,mm=func_incal2(tt)
    return(xx-x2,yy-y2,zz-z2,mm) 
```
